# Importing necessary libraries and modules

```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from matplotlib import pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
import string
```

# Open text_file in read mode and read it using read function

```python
handler = open(r"C:\Users\fateh\Documents\sample2.txt", 'r')
```

```python
Data = handler.read()
```

# Remove all the punctuations in text. import string to use "string.punctuation", i.e it will provide us punctuations

```python
Data_without_punctuation = Data.translate(str.maketrans('', '', string.punctuation))
```

# Now our data don't have punctuations. The next step is to split it into words.

```python
words = Data_without_punctuation.split()
lower_words = [word.lower() for word in words]
```

# We'll remove the following stop_words from our text, like is, am, are, etc.

```python
StopWords = set(nltk.corpus.stopwords.words('english'))
```

```python
words_without_stopwords = [word for word in lower_words if not word in StopWords]
```

# Here we are applying lemmitizer on our clean text to convert words and into their origin words

```python
lemmatizer = nltk.wordnet.WordNetLemmatizer()
clean_words = [lemmatizer.lemmatize(word) for word in words_without_stopwords]
```

# Make a dictionary to have a count of words in it, which appeared in text

```python
words_dictionary = {}
for word in clean_words:
    words_dictionary[word] = words_dictionary.get(word, 0) + 1
```

# Finally it's time to make a word-cloud and show it below.

```python
cloud = WordCloud( width = 400, height = 200)
wordcloud = cloud.generate_from_frequencies(words_dictionary)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
